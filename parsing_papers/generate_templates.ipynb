```python
import pandas as pd
```

```python
df = pd.read_csv('papers.csv')
```

```python
for paper in df.iloc:

    file_name = paper['title'].replace(' ', '').replace(':','').replace(',','').replace('/','').replace('?','')

    f = open('./'+file_name+'.md', 'w')

    content = [
        '---',
        '\n'
        'layout: technique',
        '\n'
        'title: '+paper['title'],
        '\n',
        'classifications:',
        '\n'
    ]

    for index, key in enumerate(df.keys()):
        if(key != 'title' and key != "venue" and key != 'doi'):
            content.append("    "+key+': \"'+str(paper[key])+'\"')
            if(index != len(df.keys())-1):
                content.append('\n')

    for line in content:
        print(line)
        f.write(line)
    
    f.close
```
