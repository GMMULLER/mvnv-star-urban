```python
import pandas as pd
```

```python
df = pd.read_csv('papers.csv')
```

```python
for paper in df.iloc:

    file_name = paper['title'].replace(' ', '').replace(':','').replace(',','').replace('/','').replace('?','')

    f = open('papers/'+file_name+'.md', 'w')

    content = [
        '---',
        '\n'
        'layout: technique',
        '\n'
        'authors: ',
        '\n',
        'key: \"key'+str(paper['identifier'])+'\"',
        '\n'
    ]

    for key in df.keys():
        content.append(key+': \"'+str(paper[key])+'\"')
        content.append('\n')

    content += [
        '\n',
        '---',
        '\n',
        '## Abtract'
    ]

    for line in content:
        print(line)
        f.write(line)
    
    f.close
```
